```python
data_upsampled = pd.concat([category_0, category_1_oversampled], axis=0)
```

```python
X_new = data_upsampled.drop(columns = ['STROKE'], axis=1)
y_new = data_upsampled['STROKE']
```

```python
from sklearn.model_selection import train_test_split
X_train_new, X_test_new, y_train_new, y_test_new = train_test_split(X_new, y_new, test_size=0.2, random_state=1989)
print('Shape of X_train:', X_train_new.shape)
print('Shape of X_test:', X_test_new.shape)
print('Shape of y_train:', y_train_new.shape)
print('Shape of y_test:', y_test_new.shape)
```

```python
X_train_num_new = X_train_new.select_dtypes(include=np.number)
X_test_num_new = X_test_new.select_dtypes(include=np.number)

X_train_cat_new = X_train_new.select_dtypes(exclude=np.number)
X_test_cat_new = X_test_new.select_dtypes(exclude=np.number)

X_train_num_new.shape, X_train_cat_new.shape, X_test_num_new.shape, X_test_cat_new.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler().fit(X_train_num_new)
X_train_num_scaled_new = scaler.transform(X_train_num_new)
X_test_num_scaled_new = scaler.transform(X_test_num_new)
```

```python
X_train_num_scaled_df_new = pd.DataFrame(X_train_num_scaled_new, columns = X_train_num_new.columns, index=X_train_num_new.index) 
X_test_num_scaled_df_new = pd.DataFrame(X_test_num_scaled_new, columns = X_test_num_new.columns, index=X_test_num_new.index)
```

```python
ordinal_encoding = {
    'GENHLTH': {'': 0, 'Poor': 1, 'Fair': 2, 'Excellent': 3, 'Good': 4, 'Very good': 5},
    'DIABETES': {'': 0, 'No': 1, 'Yes': 2, 'No, prediabetes': 3, 'Yes, pregnancy': 4},
    'SNUS_CHE_TOBAC': {'': 0, 'Not at all': 3, 'Every day': 2, 'Some days': 1},
    'RACE': {'Native Hawaiian or other Pacific Islander only': 1, 
             'American Indian or Alaskan Native only': 2,
             'Multiracial': 3, 'Asian': 4, 'Black': 5, 'Hispanic': 6, 'White': 7},
    'AGE': {'18 to 24': 1, '25 to 34': 2, '35 to 44': 3, '45 to 54': 4, '55 to 64': 5, '65 or older': 6},
    'BMI_GROUP': {'Underweight': 1, 'Normal Weight': 2, 'Overweight': 3, 'Obese': 4},
    'SMOKING_STATUS': {'': 0, 'Smokes some days': 1, 'Smokes every day': 2, 'Former smoker': 3, 'Never smoked': 4}
}


X_train_cat_encoded_new = X_train_cat_new.replace(ordinal_encoding)
X_test_cat_encoded_new = X_test_cat_new.replace(ordinal_encoding)
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(drop='first')
X_train_cat_ohe_new = encoder.fit_transform(X_train_cat_encoded_new)
X_test_cat_ohe_new = encoder.transform(X_test_cat_encoded_new)
```

```python
X_train_cat_df_new = pd.DataFrame(X_train_cat_ohe_new.toarray(), columns=encoder.get_feature_names_out(X_train_cat_encoded_new.columns), index=X_train_cat_encoded_new.index)
X_test_cat_df_new = pd.DataFrame(X_test_cat_ohe_new.toarray(), columns=encoder.get_feature_names_out(X_test_cat_encoded_new.columns), index=X_test_cat_encoded_new.index)
```

```python
X_train_processed_new = pd.concat([X_train_num_scaled_df_new, X_train_cat_df_new], axis=1)
X_test_processed_new = pd.concat([X_test_num_scaled_df_new, X_test_cat_df_new], axis=1)
print(X_train_processed_new.shape)
print(X_test_processed_new.shape)
```
